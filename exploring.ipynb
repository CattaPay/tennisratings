```python
dat = pd.read_csv("createddata/cam_predictions.csv")
```

```python
dat["Predictions"] = dat.apply(lambda row: 2 - row.Predictions, axis = 1)
```

```python
dat["Correct"] = dat.apply(lambda row: int(row.Match_Winner == row.Predictions), axis = 1)
```
